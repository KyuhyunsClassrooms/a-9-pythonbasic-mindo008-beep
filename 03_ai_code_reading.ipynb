## 5. 간단한 텍스트 판별 코드 읽기

아래 코드는 문장 안에 긍정 단어가 있으면 `"긍정"` 으로 바꾸는 예시입니다.


````python
positive_words = ["좋", "최고", "재미"]
text = "이 영화는 정말 재미있다"
label = "부정"

for word in positive_words:
    if word in text:
        label = "긍정"

print(label)
````

````python
negative_words = ["싫", "최악", "지루", "별로"]
text = "이 영화는 정말 최악이다"
label = "긍정"  

for word in negative_words:
    if word in text:
        label = "부정"  

print(label)
       
````

## 7. 미니 읽기 연습

아래 코드는 목록에서 하나의 주제를 고르고,  
난이도에 따라 다른 문장을 출력합니다.


````python
import random

prompts = [
    {"topic": "apple", "difficulty": "easy"},
    {"topic": "robot", "difficulty": "medium"},
    {"topic": "space", "difficulty": "easy"}
]

selected = random.choice(prompts)

if selected["difficulty"] == "easy":
    print("쉬운 주제:", selected["topic"])
else:
    print("보통 주제:", selected["topic"])
````
