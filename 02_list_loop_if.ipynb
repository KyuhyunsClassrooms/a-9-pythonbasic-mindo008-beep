## 3. `for` 는 리스트의 값을 하나씩 꺼내서 반복한다

아래 코드를 실행하기 전에  
`animal` 이 어떤 순서로 바뀔지 먼저 생각해 보세요.


```python
animals = ["고양이", "강아지", "토끼"]
num=0

for animal in animals:
    print(animal)
    print("hi")
    print(num)
    num=num+1
```

### 읽는 방법
`for animal in animals:` 는  
`animals` 안의 값을 하나씩 꺼내서 `animal` 에 넣으며 반복한다고 읽으면 됩니다.


## 4. `if` 는 조건이 참일 때만 실행한다

아래 코드는 80점 이상인 점수가 몇 개인지 세는 코드입니다.


```python
scores = [72, 85, 91, 60]
count = 0

for score in scores:
    if score >= 80:
        count = count + 1

print(count)
```

## 6. 반복문과 조건문을 함께 읽기


```python
foods = ["김밥", "라면", "비빔밥", "우동"]
long_name_count = 0

for food in foods:
    if len(food) >= 3:
        long_name_count = long_name_count + 1

print(long_name_count)
```
